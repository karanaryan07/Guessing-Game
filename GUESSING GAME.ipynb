```python
#Guessing Game
print('Enter the range in which you want to guess : (a,b) \n for example if you want to guess between 20 and 50')
#a = int(input())
a = int(input("Enter the 1st number :"))
b = int(input("Enter the 2nd number :"))
from random import randint
num = randint(a,b)
org_num = num


lower = num - 10
upper = num + 10


user_input = int(input('Enter your guess :'))
x = 1

if user_input < a or user_input > b:
    print("Out of Bounds")
else:    
    if user_input == num:
        print('You have guessed the right number : {}'.format(num))

    if user_input > lower and user_input < upper:
        print("WARM")
        while True:
            user_input_1 = int(input('Enter your guess :'))
            num2 = user_input

            if user_input_1 == org_num:
                print('You have guessed the right number : {}'.format(num))
                break

            elif (user_input_1 > num2 and user_input_1 < num) or (user_input_1 < num2 and user_input_1 > num):
                print("WARMER")
                x = x+1
                user_input = user_input_1

            elif (user_input_1 > num2 and user_input_1 > num) or (user_input_1 < num2 and user_input_1 < num):
                print("COLDER")
                x = x+1
                user_input = user_input_1

    elif user_input <= lower or user_input >= upper:
        print("COLD")
        while True:
            user_input_1 = int(input('Enter your guess :'))
            num2 = user_input


            if user_input_1 == org_num:
                print('You have guessed the right number : {}'.format(num))
                break

            elif (user_input_1 > num2 and user_input_1 < num) or (user_input_1 < num2 and user_input_1 > num):
                print("WARMER")
                x = x+1
                user_input = user_input_1
            elif (user_input_1 > num2 and user_input_1 > num) or (user_input_1 < num2 and user_input_1 < num):
                print("COLDER")
                x = x+1
                user_input = user_input_1



    print("The number of guesses you took is : {}".format(x))
```
